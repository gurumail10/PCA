```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA 
```

```python
pca=PCA()
```

```python
oli=datasets.fetch_olivetti_faces()
```

```python
oli.keys()
```

```python
fig=plt.figure(figsize=(8,8))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1)
    ax.imshow(oli.images[i],cmap=plt.cm.bone)

plt.show()    
```

```python
x_transform=pca.fit_transform(oli.data)
```

```python
eigen_val=pca.components_
eigen_val.shape
```

```python
pca.fit(oli.data)
pca.explained_variance_.shape
```

```python
k=0
current_variance=0
total=sum(pca.explained_variance_)

while current_variance/total<=0.95:
  current_variance += pca.explained_variance_[k]
  k=k+1
k  
  
```

```python
pca=PCA(n_components=k,whiten=True)
data_transform=pca.fit_transform(oli.data)
```

```python
data_transform.shape
```

```python
eigen_val=pca.components_
eigen_val.shape
```

```python
eigen_vec=eigen_val.reshape(123,64,64)
eigen_vec.shape
```

```python
fig=plt.figure(figsize=(8,8))

for i in range(64):
  ax=fig.add_subplot(8,8,i+1)
  ax.imshow(eigen_vec[i])
  
plt.show()  
  
```
